```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
URL = 'https://gaana.com/songs'
```

```python
page=requests.get(URL).text
```

```python
soup=BeautifulSoup(page,'lxml')
soup
```

```python
formula = soup.find("ul",class_ = "_row list_data")
formula.text
```

```python
singer = soup.find_all("div",class_="_art t_over")
singer
```

```python
singer_name = []
for i in singer:
    name=i.text.strip()
    singer_name.append(name)
print(singer_name)
```

```python
song_name=soup.find_all("a",class_="_tra t_over _plyCr")
song_name
```

```python
song = []
for i in song_name:
    name=i.text.strip()
    song.append(name)
```

```python
print(song)
print(len(song))
```

```python
df=pd.DataFrame({'song_name' : song, 'singer' : singer_name })
df
```

```python
df.to_csv('song_data')
```
